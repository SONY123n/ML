```python
x=df.drop('mpg',axis=1)
y=df['mpg']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
regression_model = LinearRegression()
regression_model
```

```python
regression_model.fit(x_train,y_train)
```

```python
for i,col_name in enumerate(x_train.columns):
    print(f"the coefficient for {col_name} is {regression_model.coef_[i]}")
```

```python
for i,col_name in enumerate(x_train.columns):
    print(i,col_name)
```

```python
#enumerate >> 
```

```python
from sklearn.metrics import r2_score
y_pred_linear = regression_model.predict(x_test)
r2_score(y_test,y_pred_linear)
```

```python
#ridge regression >> overfiiting 

from sklearn.linear_model import Ridge
ridge_regression_model=Ridge(alpha = 100)
ridge_regression_model.fit(x_train,y_train)
for i,col_name in enumerate(x_train.columns):
    print(f"the coefficient for {col_name} is {ridge_regression_model.coef_[i]}")
```

```python
from sklearn.metrics import r2_score
y_pred_linear = ridge_regression_model.predict(x_test)
r2_score(y_test,y_pred_linear)
```

```python
#lasso regresion >> to know the feature scale

from sklearn.linear_model import Lasso
lasso_regression_model=Lasso(alpha = 0.1)
lasso_regression_model.fit(x_train,y_train)
for i,col_name in enumerate(x_train.columns):
    print(f"the coefficient for {col_name} is {lasso_regression_model.coef_[i]}")
```

```python
from sklearn.metrics import r2_score
y_pred_linear = lasso_regression_model.predict(x_test)
r2_score(y_test,y_pred_linear)
```

```python
from sklearn.linear_model import ElasticNet
```

```python
elasticnet_regression_model=ElasticNet(alpha = 1,l1_ratio=0.5)
elasticnet_regression_model.fit(x_train,y_train)
for i,col_name in enumerate(x_train.columns):
    print(f"the coefficient for {col_name} is {elasticnet_regression_model.coef_[i]}")
```

```python
from sklearn.metrics import r2_score
y_pred_linear = elasticnet_regression_model.predict(x_test)
r2_score(y_test,y_pred_linear)
```

```python
from sklearn.linear_model import LassoCV  #cv is called here is crossvalidation
lassocv = LassoCV(cv =5 )
lassocv.fit(x_train,y_train)
y_pred=lassocv.predict(x_test)
r2_score(y_test,y_pred)
```
