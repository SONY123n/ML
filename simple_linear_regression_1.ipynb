```python
data=pd.DataFrame(diabetes.data,columns =diabetes . feature_names)
```

```python
data['target'] = diabetes.target
```

```python
data
```

```python
#data preparation
#divide the data into x and y

```

```python
#simple linear regression
#x is nothing the independent variable or the perdictor
#y is nothing the target variable or the outcome variable
x=data[['bmi']]
y=data[['target']]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.20,random_state = 111)
#using random state that means the values which are generated they won't change when u excute the sam eline of code many times
x_train
```

```python
y_train
```

```python
#scaling  
# for scaling here fit (calculate the parameters like mu and sigma) and transform (apply the parameters to caluclate the zscore) both the parameters are used for the tarin data and only transform is used for the test data
```

```python
#model training
#same like the scaling we have fit and predict for the ML same as the scaling the both fit-predict is used for the train data and only predict is used for the test data
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
model=LinearRegression()
```

```python
model
```

```python
model.fit(x_train,y_train)
```

```python
model.coef_ #this is m
```

```python
model.intercept_ # this is c
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
#visualised the result
plt.scatter(x_test,y_test,color='#e0218a',label="Actual data")
plt.plot(x_test,y_pred,color="green",linewidth = 3 ,label ="linear regression line")
plt.xlabel("BMI")
plt.ylabel("one yera preogression-target")
plt.title("my first linear regression model")
plt.legend()
plt.show()
```
