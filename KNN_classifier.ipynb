```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_classification
x,y=make_classification(n_samples =1000,n_features=3,n_redundant =0,n_classes =2,random_state=1)
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=1,test_size=0.3)
```

```python
x_train.shape,x_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier()
```

```python
clf
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
accuracy_score(y_test,y_pred)
```

```python
classification_report(y_test,y_pred)
```

```python
param_grid={"n_neighbors" : [3,5,6,7,10,12,15],
            "algorithm" :["ball_tree","brute","kd_tree"],
            "leaf_size":[20,30,40,50]
           }
```

```python
param_grid
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator=clf,param_grid=param_grid,cv=5,verbose=3)
grid
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
best_model = grid.best_estimator_
```

```python
y_pred_tune=best_model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred_tune)
```
