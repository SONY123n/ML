```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix


import warnings
warnings.filterwarnings('ignore')
```

```python
x,y=make_classification(n_samples=1000,n_features=20,n_classes=2,random_state=1)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state= 1)
```

```python
 classifier = GradientBoostingClassifier()
```

```python
classifier
```

```python
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {"n_estimators" : [50,100,200],
              "learning_rate" : [0.001,0.01,1.5,2,2.5],
              }

gd =  GradientBoostingClassifier()

clf = GridSearchCV(estimator = gd,param_grid=param_grid,cv=5,verbose=3)
clf
```

```python
clf.fit(x_train,y_train)
```

```python
y_p = clf.predict(x_test)
```

```python
y_p
```

```python
clf.best_params_
```

```python
clf.best_score_
```

```python
best_model = clf.best_estimator_
y_pd = best_model.predict(x_test)
```

```python
accuracy_score(y_test,y_pd)
```

```python
#ada boost regressor

from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score
```

```python
x,y=make_regression(n_samples=1000,n_features=2,noise=10,random_state=1)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state= 1)
```

```python
regressor = GradientBoostingRegressor()
regressor
```

```python
regressor.fit(x_train,y_train)
```

```python
y_pred = regressor.predict(x_test)
r2_score(y_test,y_pred)
```
