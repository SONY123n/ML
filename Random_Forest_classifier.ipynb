```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['time']=encoder.fit_transform(df['time'])
```

```python
x=df.drop('time',axis=1)
y=df['time']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
#handling missing value
#data encoding
#feature sacling
```

```python
from sklearn.impute import SimpleImputer # for missing values
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler

#pipeline >>a squence of data transformation 
#column transformation >> groups all the pipeline steps for eacg of the columns

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
df
```

```python
cat_cols = ['sex','smoker','day']
num_cols =['total_bill','tip','size']
```

```python
#feature engineering automachine using pipeline and column transformer

num_pipeline = Pipeline(steps =[('imputation',SimpleImputer(strategy = 'median')),
                                ('sacling',StandardScaler())])
cat_pipeline =Pipeline(steps =[('imputation',SimpleImputer(strategy = 'most_frequent')),
                                ('encoding',OneHotEncoder())])
```

```python
preprocessor = ColumnTransformer([('num_pipeline',num_pipeline,num_cols),
                   ('cat_pipeline',cat_pipeline,cat_cols)])
```

```python
x_train = preprocessor.fit_transform(x_train)
x_test = preprocessor.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
```

```python
rf
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.model_selection import GridSearchCV
params={'max_depth':[1,2,3,10,20,50],
        'n_estimators':[50,100,200,300,400,1000],
        'criterion':['gini','entropy']}
```

```python
params
```

```python
clf=GridSearchCV(rf,param_grid=params,cv=5,scoring='accuracy',verbose=3)
```

```python
clf
```

```python
clf.fit(x_train,y_train)
```
