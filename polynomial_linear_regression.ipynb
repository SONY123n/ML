```python
np.random.seed(1)
```

```python
x=2*np.random.rand(100,1)
```

```python
y=4+3*x + 1.5*x**2 + np.random.rand(100,1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```
